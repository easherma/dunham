```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
# read in dataframe
df = pd.read_csv('dunham_join.csv')
```

```python
"""
Loop through data and count two things:

Length of stay = number of consecutive days in same city, call it a single 'trip' (may indicate success in an area)

# of returns to an city = number of trips to a city ( may indicate prefrence )
"""

def compare_previous(df, column):
    """
    compare previous row for change in column value
    """
    return df[column].eq(df[column].shift())

def count_duration(trips):
    """
    count number of rows column value stays the same
    """
    return sorted(trips, key= lambda x: x[1], reverse=True)

def count_returns(trips):
    counts = Counter(elem[0] for elem in trips)
    returns = counts.most_common()
    return returns

def trip_duration_count(df, column):
    """
    Initial loop through data counting trips and their date ranges
    
    """
    count = 0
    results = []
    for i, v in df.iterrows():
        min_date = df['Date'][i]
        lat = v['lat:1']
        lng = v['lng']
        print(lat)
        print(lng)

        if v['same'] != False:
            count += 1
            max_date = df['Date'][count]
            
#             print(i, v['Date'], v[column])
        else:
#             result = {df[column][i-1]: count}
            max_date = df['Date'][i-1]
            min_date = df['Date'][i-count]
            
            result = (df[column][i-1] , count, min_date, max_date, df['lat:1'][i-1], df['lng'][i-1] )
            # just one entry is 1
            if result[1] == 0:
#                 result[df[column][i-1]] = 1
                result = (df[column][i-1] , count+1, min_date, max_date, df['lat:1'][i-1], df['lng'][i-1])
#             print(result)
            #index of trips dict set here
#             results[len(results)] = result
            results.append(result)
            count = 0
#         print(v['same'], min_date, max_date)
    return results
    #     print(i+1, v['City, Country'], v['same'])
    

df['same'] = compare_previous(df, 'City, Country')
#change first entry value to true
df['same'][0] = True

trips = trip_duration_count(df, 'City 1')
durations = count_duration(trips)
returns = count_returns(trips)
```
