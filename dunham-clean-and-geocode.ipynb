```python
import pandas as pd
import numpy as np
from collections import Counter
import geocoder
```

```python
# read in dataframe
df = pd.read_excel("CURA Dunham 1950-53 .xlsx")
```

```python
# lowercase columns
df.columns = [x.lower() for x in df.columns]
df.columns = [x.strip() for x in df.columns]
df.columns = [x.replace(" ", "_") for x in df.columns]
```

```python
#df = pd.read_csv('g2.csv')
```

```python
df['lat'] = None
df['lng'] = None
df['geocoded_location'] = None
```

```python
# this is where we geocode
for i, v in df.iterrows():

    if pd.isnull(df['lat'][i]):
        print(str(v['city_1']) + "," + str(v['country_1']))
        try:
            g = geocoder.google(str(v['city_1']) + "," + str(v['country_1']))
            df['lat'][i] = g.lat
            df['lng'][i] = g.lng
            df['location'][i] = g.location
        except Exception as e:
            print(e)
            df['lat'][i] = None
            df['lng'][i] = None
            df['geocoded_location'][i] = None
        
```
