```python
import pandas as pd
import numpy as np
from collections import Counter
import geocoder
```

```python
# read in dataframe
df = pd.read_excel("CURA Dunham 1950-53 .xlsx")
```

```python
# lowercase columns
df.columns = [x.lower() for x in df.columns]
df.columns = [x.strip() for x in df.columns]
df.columns = [x.replace(" ", "_") for x in df.columns]
```

```python
g = geocoder.google(df['city_1'][0] + "," + df['country_1'][0])
```

```python
for i, v in df.iterrows():
#     print(str(v['city_1']) + "," + str(v['country_1']))
    if df['lat'][i] == None:
        try:
            g = geocoder.google(str(v['city_1']) + "," + str(v['country_1']))
            df['lat'][i] = g.lat
            df['lng'][i] = g.lng
        except Exception as e:
            print(e)
            df['lat'][i] = None
            df['lng'][i] = None
        
```

```python
df.to_csv("g2.csv")
```

```python
"""
Loop through data and count two things:

Length of stay = number of consecutive days in same city, call it a single 'trip' (may indicate success in an area)

# of returns to an city = number of trips to a city ( may indicate prefrence )
"""

def compare_previous(df, column):
    """
    compare previous row for change in column value
    """
    return df[column].eq(df[column].shift())

def count_duration(trips):
    """
    count number of rows column value stays the same
    """
    return sorted(trips, key= lambda x: x[1], reverse=True)

def count_returns(trips):
    counts = Counter(elem[0] for elem in trips)
    returns = counts.most_common()
    return returns

def trip_duration_count(df, column):
    """
    Initial loop through data counting trips and their date ranges
    
    """
    count = 0
    results = []
    for i, v in df.iterrows():
        min_date = df['date'][i]
        lat = v['lat']
        lng = v['lng']
        print(lat)
        print(lng)

        if v['same'] != False:
            count += 1
            max_date = df['date'][count]
            
#             print(i, v['Date'], v[column])
        else:
#             result = {df[column][i-1]: count}
            max_date = df['date'][i-1]
            min_date = df['date'][i-count]
            
            result = (df[column][i-1] , count, min_date, max_date, df['lat'][i-1], df['lng'][i-1] )
            # just one entry is 1
            if result[1] == 0:
#                 result[df[column][i-1]] = 1
                result = (df[column][i-1] , count+1, min_date, max_date, df['lat'][i-1], df['lng'][i-1])
#             print(result)
            #index of trips dict set here
#             results[len(results)] = result
            results.append(result)
            count = 0
#         print(v['same'], min_date, max_date)
    return results
    #     print(i+1, v['City, Country'], v['same'])
    

df['same'] = compare_previous(df, 'city_1')
#change first entry value to true
df['same'][0] = True

trips = trip_duration_count(df, 'city_1')
durations = count_duration(trips)
returns = count_returns(trips)
```

```python
durations_df = pd.DataFrame.from_records(durations, columns=['city_1', 'duration_days', 'min_date', 'max_date', 'lat', 'lng'] )
returns_df = pd.DataFrame.from_records(returns, columns=['city_1', 'returns'])

durations_df.to_csv('durations.csv')
returns_df.to_csv('returns.csv')
```

```python
df_duration_merge = pd.merge(df, durations_df, left_on = 'city_1', right_on = 'city_1')
```
